```python
import piexif
from fractions import Fraction
```

```python
def get_exif_data_key(file_path):
    tags = {'f_number': piexif.ExifIFD.FNumber,
            'iso': piexif.ExifIFD.ISOSpeedRatings, 
            'exposure_time': piexif.ExifIFD.ExposureTime}
    
    exif_dict = piexif.load(file_path)['Exif']
    data = {}
    for k,v in tags.items():
        if v in exif_dict:
            if k == 'f_number':
                v_tup = exif_dict[v]
                try:
                    f_number = str(Fraction(v_tup[1], v_tup[0])).replace('/', '_')
                except ZeroDivisionError:
                    f_number = "UNKNOWN"
                
                data[k] = f_number
                continue
            if k == 'exposure_time':
                v_tup = exif_dict[v]
                e_time = str(Fraction(v_tup[0], v_tup[1])).replace('/', '_')
                data[k] = e_time
                continue
            data[k] = exif_dict[v]
    return f"{data['f_number']}-{data['iso']}-{data['exposure_time']}"
    

#get_exif_data_key("D:/astro/Moon_09_06_2022/Raws/DSC01076.tiff")  
```

```python
import glob
dir_root = "D:/astro/Moon_12-13_09_2022"
files = glob.glob(f"{dir_root}/tiff/*.tiff")
print(files)
files_data_by_key = {}
for file in files:
    k = get_exif_data_key(file)
    k_files = files_data_by_key.get(k, [])
    k_files.append(file)
    files_data_by_key[k] = k_files

print(files_data_by_key)

files_data=[]
for file in files:
    k = get_exif_data_key(file)
    files_data.append({'file': file, 'key': k})
```

```python
import pandas as pd
files_data_df = pd.DataFrame.from_records(files_data)
files_data_df
```

```python
files_data_df.groupby(['key']).count()
```

```python
import pathlib
import shutil

dir_path = pathlib.PurePath(dir_root)
for key, key_files in files_data_by_key.items():
    # create directory dir_root/key/raw copy key_files in it.
    key_dir = dir_path / key
    key_dir_raw = key_dir / "tiff"
    pathlib.Path(key_dir_raw).mkdir(parents=True, exist_ok=True)
    for key_file in key_files:
        shutil.copy2(key_file, key_dir_raw)
```

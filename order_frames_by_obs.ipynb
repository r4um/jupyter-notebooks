```python
import astropy
from astropy.io import fits
import os
import glob
import pandas as pd
import numpy as np
from humanfriendly import format_timespan
from datetime import datetime
from xisf import XISF
import dateutil
```

```python
astro_files = glob.glob("D:\\astro\\iTelescope_T71_0007_Iris\\proc\\*.xisf")

astro_files_date_obs = {}
for astro_file in astro_files:
    #print(astro_file)
    astro_file_xisf = XISF(astro_file)
    astro_file_xisf_date_obs = dateutil.parser.parse(
        astro_file_xisf.get_images_metadata()[0]['FITSKeywords']['DATE-OBS'][0]['value']
    )
    astro_files_date_obs[astro_file] = astro_file_xisf_date_obs

print(astro_files_date_obs)
astro_files_date_obs=sorted(astro_files_date_obs, key=astro_files_date_obs.get)
#print(astro_files_date_obs)
```

```python
import pathlib
import shutil

out_dir_root = "D:\\astro\\iTelescope_T71_0007_Iris\\blink"
out_dir_path = pathlib.PurePath(out_dir_root)
pathlib.Path(out_dir_path).mkdir(parents=True, exist_ok=True)
i = 0
for i, astro_file in enumerate(astro_files_date_obs):
    new_file =  f"Blink_{i:04}.xisf"
    new_file_path = out_dir_path / new_file
    print(astro_file, " -> ", new_file)
    shutil.copy(astro_file, new_file_path)
```
